```python
hag = Graph.Read_GraphML('watched_pages2/Rubeus_Hagrid.comment_network.graphml')
```

```python
vs = [v for v in hag.vs]
```

```python
users = [v['user'] for v in hag.vs]
```

```python
user_map = {}
```

```python
for v in vs:
    try:
        ids = user_map[v['user']]
    except KeyError:
        ids = []
    ids.append(v.index)
    user_map[v['user']] = ids
```

```python
user_map.values()
```

```python
mapping = []
for v in vs:
    mapping.append(user_map[v['user']][0])
```

```python
hag.contract_vertices(mapping, 'first')
```
